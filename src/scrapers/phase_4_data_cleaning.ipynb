```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
df=pd.read_csv(r"C:\Users\JAY LODHA\web-scraping-project\data\raw\books_scraping_dataset_updated_phase_3.csv")
df.info()
```

```python
df.describe()
df.head()
```

```python
df["Rating"]=df['Rating'].replace(["One","Three","Two","Four","Five"],[1,3,2,4,5],inplace=True).astype(int)
```

```python
df.duplicated().value_counts()
```

```python
df.isnull().value_counts()
```

```python
df['Num Reviews'].value_counts()
df['Tax'].value_counts()
df['Product Type'].value_counts()
```

```python
df['Price (Excl Tax)'] = df['Price (Excl Tax)'].str.replace('£', '').astype(float)
df['Price (Incl Tax)'] = df['Price (Incl Tax)'].str.replace('£', '').astype(float)
df.iloc[:,1:3]
```

```python
norm=MinMaxScaler()
df[['Price (Excl Tax)','Price (Incl Tax)','Rating', 'Stock Count']] = norm.fit_transform(df[['Price (Excl Tax)','Price (Incl Tax)','Rating', 'Stock Count']])
```

```python
df.drop(['Tax','Num Reviews','Product Type'],axis=1,inplace=True)
df.head()
```

```python
df.to_csv(r"C:\Users\JAY LODHA\web-scraping-project\data\processed/books_scraping_data_phase_3.csv", index=False, encoding="utf-8-sig")
```
