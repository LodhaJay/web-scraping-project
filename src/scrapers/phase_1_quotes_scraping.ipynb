```python
#Import Library 
import requests
from bs4 import BeautifulSoup
import pandas as pd
import os
```

```python
#request to Webpage
url="https://quotes.toscrape.com"
response=requests.get(url)
print(response.status_code)
```

```python
#Parse the HTML content
soup=BeautifulSoup(response.text,'html.parser')
```

```python
#List for data
quotes_list=[]
authors_list=[]
tags_list=[]
```

```python
#Extracting Data
for quote_block in soup.find_all('div',class_='quote'):
    quote_text=quote_block.find('span',class_='text').text
    author=quote_block.find('small',class_='author').text
    tags=[tag.text for tag in quote_block.find_all('a',class_='tag')]
    
    # Append data to lists
    quotes_list.append(quote_text)
    authors_list.append(author)
    tags_list.append(tags)
```

```python
#Printing data
print("quotes: ") #printing quotes
for quote in quotes_list:
    print(quote)
    print()
print("Authors: ",authors_list) # Print authors
print("Tags: ",tags_list) # Print tags
```

```python
#save to CSV and JSON
data=pd.DataFrame({
    "Quotes":quotes_list,
    "Authors":authors_list,
    "Tags":[", ".join(tags) for tags in tags_list]
})
data.to_csv("CSV/quotes_dataset_phase_1.csv", index=False, encoding="utf-8-sig")
data.to_json("CSV/quotes_dataset_phase_1.json", orient="records", indent=4, force_ascii=False)
```
